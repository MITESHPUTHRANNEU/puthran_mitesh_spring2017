## Finding top 20 people in Enron who recieved the most emails 

```python
import os
import glob
import email

frequency={}
path = r'C:\Users\mites\Desktop\PYTHON\Midterm\Data\Enron\maildir'
for infile in glob.glob( os.path.join(path, '*') ): #using glob.glob to get all the files
    for x in glob.glob(os.path.join(infile, '*')):  #traversing inside the folders
        subpath = glob.glob(os.path.join(x,'*'))
        for y in subpath:
            if y[-1].isdigit():  # to make sure that the files are of type number
                #print(y)
                with open(y,'r') as myfile:
                    data=myfile.read()
                    #print(data)
                    msg = email.message_from_string(data) 
                    p = str(msg['X-To']).split()  #getting the name of the person from 'X-To' section of the emails
                    for w in p:
                        if w not in frequency:
                            frequency[w] = 1
                        else:
                            frequency[w] += 1

freqwords=sorted(frequency.items(),  key=lambda x: x[1], reverse=True) # using sorted function to sort the words 
print("Top 20 people who recieved the most emails in Enron: ")
print(freqwords[:20])
```

## Exporting the data to a CSV file

```python
import csv    # this part of the code is used to put the data into the csv file.
n=0
list(n)
title = ['Name-Rank-No.ofMails']   # to display the first row of the csv file
with open('recieve.csv', 'w',newline="") as out:
    writer = csv.writer(out)
    for x in title:
        writer.writerow(x.split("-")) # splits the title into different colums 
    for e in freqwords:
        n=n+1
        writer.writerow([e,n]) #printing the data into the csv files
```

## Plotting graph for number of emails recieved by top 20 recievers

```python
import matplotlib.pyplot as plt
import csv

x = []   # making empty lists
y = []

with open('recieve.csv','r') as csvfile:
    plots = csv.reader(csvfile, delimiter=',')
    next(csvfile)
    for row in plots:
        x.append(int(row[1]))  #row[1] is the rank column in the excel file
        y.append(int(row[2]))  #row[2] is the frequency column in the excel file

plt.loglog(x,y, label="Emails Recieved",color='g',linewidth=2)  # plotting the graph for number of emails against its rank using log-log
plt.xlabel('Rank of Email number') #X-axis label
plt.ylabel('Number of emails recieved') #Y-axis label
plt.grid(True)
plt.legend()
plt.show()
```
