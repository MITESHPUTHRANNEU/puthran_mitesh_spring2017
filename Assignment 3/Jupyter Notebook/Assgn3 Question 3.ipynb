```python
import pandas as pd
from pandas import Series, DataFrame
a = pd.read_csv('C:\\Users\mites\Python\DataAnalysis4Python_Spring17\Assignments\Assignment 3\Data\\cricket_matches.csv')
a[:10]
```

```python
frame = DataFrame(a, columns=['home', 'winner', 'innings1', 'innings1_runs','innings2', 'innings2_runs'])
frame[:10]
```

```python
frame2 = frame.loc[frame['home'] == frame['winner']]
frame2[:10]
```

```python
frame2['scores'] = frame2['innings1_runs'].where(frame2['innings1'] == frame2['home'], frame2['innings2_runs'])
```

```python
frame3=frame2.groupby('home')['scores'].mean().head(5)
frame3
```

```python
frame3.to_csv('scores.csv')
```
